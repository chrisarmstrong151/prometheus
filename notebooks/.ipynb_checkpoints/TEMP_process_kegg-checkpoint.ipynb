# Process KEGG KGML files

We will process the KGML files to produce directed graphs.  These graphs can then be queried to ascertain topological relationships within the various pathways.

# Libraries

```python
# Libraries


# Biopython for KGML parsing.
from Bio.KEGG.KGML.KGML_parser import read as kegg_read

# Graphs.
import networkx as nx

# File operations.
import os
import re
```

# Pipeline

### Class definition.

```python
# The class holds all pathways and any relevant
# metrics used to compare them.
class pathway_super:
    
    # Initialization.
    def __init__(self):
        
        # Blank the KGML directory.
        self.kgml_directory = None
    
#     # Create the basic shared names across
#     # pathways graph.
#     def basic_shared_names_across_pathways(self, layout = 'most_connected'):
        
#         # layout: most_connected or heaviest_connected
        
#         # layout determines the type of graph made.
#         print('here')
        
    
#     # Initialize the metrics.
#     def initialize_metrics(self):
        
#         # An attribute to hold the metrics.
#         self.metrics = {}
        
#         # Entry connections.
#         self.metrics['entry_connections'] = {
#             'entry': [],
            
#         }
    
    # A metric to establish Entry connections.
    def metric_entry_connections(self):
        
        # Read the self.pairwise_shared_entries
        # attribute and calculate connectivities.
        self.metric_pathway_to_pathway_connectivities = {}
        
        # Save the overall connectivity for each entry.
        self.metric_overall_entry_connectivities = {}
        
        # Go over every edge pair and its connections.
        for edge_pair, connections in self.pairwise_shared_entries.items():
            
            # Split the edge pair name.
            edge_pair_name_split = edge_pair.split('-')
            
            # Are the keys available in overall_entry_connectivities?
            if edge_pair_name_split[0] not in self.metric_overall_entry_connectivities:
                self.metric_overall_entry_connectivities[edge_pair_name_split[0]] = 0
            
            if edge_pair_name_split[1] not in self.metric_overall_entry_connectivities:
                self.metric_overall_entry_connectivities[edge_pair_name_split[1]] = 0
            
            # How many shared entries?
            n_shared_entries = len(connections)
            
            # Set the metric and sub-metric if necessary.
            if n_shared_entries > 0:
                self.metric_pathway_to_pathway_connectivities[(edge_pair_name_split[0], edge_pair_name_split[1])] = n_shared_entries
                self.metric_overall_entry_connectivities[edge_pair_name_split[0]] += 1
                self.metric_overall_entry_connectivities[edge_pair_name_split[1]] += 1
    
    # A metric to find shared pathways for an Entry.
    def metric_shared_pathways_for_entry(self):
        
        # Define an attribute which holds the results
        # of the Entry-Pathway comparisons.
        self.metric_entry_shared_pathways = {}
        
        # Go over every edge pair and its connections.
        for edge_pair, connections in self.pairwise_shared_entries.items():
            
            # For every Entry, add to the metric.
            for connection in connections:
                for entry in connection.split(' '):
                    
                    # Do we have this Entry already?
                    if entry not in self.metric_entry_shared_pathways:
                        self.metric_entry_shared_pathways[entry] = 0
                    
                    # Add 1 to the total.
                    
                    # NOTE: this metric captures only how many *pairs*
                    # the Entry is in, not how many unique Pathways
                    # the Entry is in.
                    self.metric_entry_shared_pathways[entry] += 1            
    
    # Find shared Entry names between pathways.
    def shared_entry_names_between_pathways(self):
        
        # Define an attribute which holds the results
        # of the pathway-pairwise comparisons.
        self.pairwise_shared_entries = {}
        
        # Get all Entry names, then do a simple set
        # intersection.
        for pathway_one_id in list(self.kgml_files.keys()):
            
            # Get all the genes for pathway one.
            pathway_one_genes = [i.name for i in self.kgml_files[pathway_one_id].genes]
            
            for pathway_two_id in list(self.kgml_files.keys())[1:]:
                
                # Get all the genes for pathway two.
                pathway_two_genes = [i.name for i in self.kgml_files[pathway_two_id].genes]
                
                # Now see which names are shared between
                # the two pathways.
                shared = set(pathway_one_genes).intersection(pathway_two_genes)
                
                # Set the attribute.
                self.pairwise_shared_entries[pathway_one_id + '-' + pathway_two_id] = shared
    
    # Load all pathways.
    def load_pathways(self):
        
        # What are all the files?
        
        # Adapted from Google AI response.
        regex_pattern = r".*\.xml$"
        
        # KGML files.
        kgml_files = {}
        
        for kgml_file in os.listdir(self.kgml_directory):
            if re.search(regex_pattern, kgml_file):
                
                # Create the key.
                temp_key = kgml_file.split('/')[-1].split('.')[0]
                
                # Create the value.
                kgml_files[temp_key] = kegg_read(open(self.kgml_directory + kgml_file, 'r'))
        
        # Set the attribute.
        self.kgml_files = kgml_files
    
    # Set the directory with the KGML files.
    def set_kgml_directory(self, directory):        
        
        # TODO: try/except for files, blah, blah, blah
        self.kgml_directory = directory
```

### Create the basic shared entities graph.

```python
G = nx.Graph()

# Subset to keys of interest...
keys_to_include = list(ps.metric_pathway_to_pathway_connectivities.keys())[0:1000]
subset_dict = {key: ps.metric_pathway_to_pathway_connectivities[key] for key in keys_to_include if key in ps.metric_pathway_to_pathway_connectivities}

for node_pair, connectivity_metric in subset_dict.items():
# for node_pair, connectivity_metric in ps.metric_pathway_to_pathway_connectivities.items():
    
    G.add_edge(node_pair[0], node_pair[1], weight = connectivity_metric)

    #G.add_edge(0,1,color='r',weight=2)
    #G.add_edge(1,2,color='g',weight=4)

#colors = nx.get_edge_attributes(G,'color').values()
weights = nx.get_edge_attributes(G,'weight').values()

pos = nx.circular_layout(G)
nx.draw(G, #pos, 
        #edge_color=colors, 
        width=list(weights),
        #with_labels=True,
        node_color='lightgreen')
```
