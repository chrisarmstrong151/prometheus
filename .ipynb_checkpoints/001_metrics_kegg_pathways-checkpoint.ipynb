# KEGG Pathway Metrics

We will characterize the system response given a set of genes using the KEGG pathways.

# Libraries

```python
# Libraries


# KGML files
from bs4 import BeautifulSoup

# globbing
from glob import glob

# Interweaving
import itertools

# Plotting
import networkx as nx
import matplotlib.pyplot as plt

# Pandas
import pandas as pd
```

```python
# Set the data folder.
data_folder = '/media/aeneas/Samsung_T5/transfer/mayur/'
```

# KGML files

First, read in all KGML files.

```python
# KGML files.
kgml_files = glob(data_folder + '/kgml_files/*.xml')

# Read them all in.
kgmls = {}

for kgml_file in kgml_files:
    
    # Get the pathway name.
    pathway_name = kgml_file.split('/')[-1].split('.')[0]
    
    # Read the file.
    with open(kgml_file, 'r') as f:
        
        read_in = ''.join([i.strip() for i in f.readlines()])
        
        # Parse.
        kgmls[pathway_name] = BeautifulSoup(read_in, 'lxml')
```

Find all possible types of relations and all possible types of entries.

```python
# Relation types.
relation_types = {}

# Entry types.
entry_types = []

# Go over every pathway.
for pathway_id in kgmls.keys():
    
    # Relations.
    for i in kgmls[pathway_id].find_all('subtype'):
    
        # Check to see if we've already
        # had this type.
        if i['name'] not in relation_types:
            relation_types[i['name']] = i['value']
    
    # Entries.
    for i in kgmls[pathway_id].find_all('entry'):
    
        # Check to see if we've already
        # had this type.
        if i['type'] not in entry_types:
            entry_types.append(i['type'])
```

# Metrics (aggregate)

Begin by getting every entity in every pathway.

```python
# Get every entity of every pathway.
entities_by_pathway = {}

# Go over every pathway.
for pathway_id in kgmls.keys():
    
    # Add the key.
    entities_by_pathway[pathway_id] = []
    
    # Now the entities for this pathway.    
    for i in kgmls[pathway_id].find_all('entry'):
        for j in i['name'].split(' '):
            entities_by_pathway[pathway_id].append(j)
```

```python
# entities_by_pathway
```

Now do pairwise comparisons between pathways.

```python
# Pairwise comparisons.
pairwise_comparisons = {}

# Keep track of how many connections each
# pathway has.
n_connections_by_pathway = {}

# Go over every pairwise combination
# of pathways.
for index_one in range(0, len(entities_by_pathway.keys())):    
    for index_two in range(index_one + 1, len(entities_by_pathway.keys())):
        
        # Define the pathways.
        first_pathway_key = list(entities_by_pathway.keys())[index_one]
        second_pathway_key =  list(entities_by_pathway.keys())[index_two]
        
        # Load the pathways.
        first_pathway = entities_by_pathway[first_pathway_key]
        second_pathway = entities_by_pathway[second_pathway_key]
        
        # Anything shared?
        shared = set(first_pathway).intersection(set(second_pathway))
        
        # Update the number of connections.
        if first_pathway_key not in n_connections_by_pathway:
            n_connections_by_pathway[first_pathway_key] = 0
        else:
            n_connections_by_pathway[first_pathway_key] += len(shared)
        
        if second_pathway_key not in n_connections_by_pathway:
            n_connections_by_pathway[second_pathway_key] = 0
        else:
            n_connections_by_pathway[second_pathway_key] += len(shared)
        
        if len(shared) > 0:
            
            # Set the key.
            pairwise_comparisons[first_pathway_key + '-' + second_pathway_key] = shared
```

Clean up the number of connections by pathway.

```python
# DataFrame.
values = []

for k, v in n_connections_by_pathway.items():
    values.append(v)

n_connections_by_pathway = pd.DataFrame(
    data = {
        'kegg_pathway_id': list(n_connections_by_pathway.keys()),
        'n_connections': values
    }
)
```

We can also look at the specifically shared entities between two different pathways.

```python
# DataFrame.
pathway_one = []
pathway_two = []
shared = []

for k, v in pairwise_comparisons.items():
    
    # Define the first and second pathways.
    first_pathway = k.split('-')[0]
    second_pathway = k.split('-')[1]
    
    for entity in v:
        pathway_one.append(first_pathway)
        pathway_two.append(second_pathway)
        shared.append(entity)

connections_across_pathways = pd.DataFrame(
    data = {
        'pathway_one': pathway_one,
        'pathway_two': pathway_two,
        'shared': shared
    }
)
```

```python
g = nx.Graph()

# Edges list.
edges_list = []

for pairing_key in pairwise_comparisons.keys():
    
    # From where to where?
    start = pairing_key.split('-')[0]
    stop = pairing_key.split('-')[1]
    
    # How many repeats?
    repeat_number = len(pairwise_comparisons[pairing_key])
    
    # Add to the edges list.
    for i in range(0, repeat_number):
        edges_list.append((start, stop))
    
g.add_edges_from(edges_list)

d = dict(g.degree)

nx.draw(g, nodelist=d.keys(), node_size=[v * 1 for v in d.values()])
plt.show()
```

# Metrics (pathway)

Suppose we start with a list of genes from the cell cycle (hsa04110), apoptosis (hsa04210), and B cell receptor signaling (hsa04662) pathways.

```python
# Genes of interest.
genes = ['CDK2', 'WEE', 'CASP3', 'CASP7', 'BTK', 'AKT']
```

We can conduct a causality analysis to see if the pathway results represent what is expected given a list of genes which have gone up or down between two states or time points.  Note that we can only infer causality for *downregulated* genes using only mRNA expression, but cannot infer causality for *upregulated* genes using only mRNA expression, as we do not have the protein data to confirm that the mRNA is actually translated.

In the future, we will incorporate orthogonal data to help with the analysis.

For now, we will establish the causality chains assuming that down- and upregulation are causal for down and upregulation of protein products.

Recall the available relationships between entities.

```python
relation_types
```

We will be only concerned with inhibition, activation, expression, and repression.

```python
# Relevant entity relations.
entity_relations = ['inhibition', 'activation', 'indirect effect', 'expression', 'repression']
```

Go over each KGML file and establish the causality chains.

```python
# Temporary helper.
entity_helper = {}

# Causality diagrams.
causality_diagrams = {}

# A dictionary to help calculate the
# true in- and out-degress later.
degrees_helper = {}

# Go over every pathway.
for pathway_id in kgmls.keys():
# for pathway_id in ['hsa04110']:

    print(pathway_id)

    # Make an entry in the causality diagrams.
    causality_diagrams[pathway_id] = {
        'digraph': '',
        'causality_diagrams': []
    }
    
    # Make an entry in the degrees helper.
    degrees_helper[pathway_id] = []
    
    # First, set the keys in the temporary helper.
    for i in kgmls[pathway_id].find_all('entry'):
        
        # Split the name up.  If it's undefined,
        # we have a group.
        split_up = i['name'].split(' ')
        
        if split_up == ['undefined']:
            
            entity_helper[i['id']] = []
            
            # Get the component IDs.
            for x in i.find_all('component'):
                entity_helper[i['id']].append(x['id'])
        
        else:
            
            # Normal name.
            entity_helper[i['id']] = split_up
    
#     import json    
#     print(json.dumps(entity_helper, indent = 4, sort_keys = True))
    
    # Now update entries that were for groups.
    entity_helper_copy = entity_helper.copy()
    
    for k, v in entity_helper.items():
        # print('+++++++++')
        # print(k)
        # print(v)
        
        if v[0][0] != 'h' and v[0][0] != 'p':
            
            # We will ignore drugs/compounds, orthologs, 
            # and glycans.
            if v[0][0] in ['c', 'd', 'k', 'g']:
                del entity_helper_copy[k]
            else:
                # print('===============')
                # print(entity_helper_copy[k])
                entity_helper_copy[k] = []
                # print(entity_helper_copy[k])
                for m in v:
                    for n in entity_helper[m]:
                        entity_helper_copy[k].append(n)
        
    # print(entity_helper_copy)
    
    # Now we can build the causality chains.
    
    # Begin by creating all of the nodes and then
    # building a NetworkX object.
    
    # Source: https://networkx.org/documentation/stable/reference/classes/digraph.html
    G = nx.DiGraph()
    
    # Relations.
    for j in kgmls[pathway_id].find_all('relation'):
        # print(j)
        # print('=====')
        # print(j.find_all('subtype'))
        # print('333333')
        
        # For now, ignore any relations which aren't
        # direct, gene-to-gene interactions.
        try:
            
            if j.find('subtype')['name'] in entity_relations:

                # Get the main and sub-type of relations.
                all_subtypes = j.find_all('subtype')

                # Define a string to associate with the edge.
                edge_string = ''

                for l in range(0, len(all_subtypes)):

                    # 0 is the main relation type, 
                    # and 1 is the relation subtype.
                    if l == 0:
                        edge_string = all_subtypes[l]['value']
                    else:
                        edge_string += all_subtypes[l]['value']

                # Add the edge.
                G.add_edge(
                    j['entry1'], j['entry2'],
                    relation = edge_string
                )

                # print(edge_string)
                
        except TypeError:
            pass
    
    # nx.draw(G)
    # plt.show()
    
    # G.graph
    
    # Make an undirected copy of the digraph.
    
    # Note that we LOSE directional information
    # by doing this.
    
    # Source: https://stackoverflow.com/a/21751571
    UG = G.to_undirected().copy()
    
    # Get the components of each sub-graph.
    # Source: https://stackoverflow.com/a/61537932
    components = [UG.subgraph(c).copy() for c in nx.connected_components(UG)]
    
    # Now that we know which nodes are in which
    # sub-graph, we need to re-construct the directional
    # graphs, keepin them separate.
    digraphs = {}
    
    for idx, UG in enumerate(components, start=1):
        
        # Add the key.
        # print(f"Component {idx}: Nodes: {UG.nodes()} Edges: {UG.edges()}")
        digraphs['-'.join(list(UG.nodes()))] = G.subgraph(UG.nodes()).copy()
        
    # print(digraphs)
    # print(UG.graph)
    
    # print('here')
    
    # Now store the causality diagrams.
    
    # Go over every node.
    for a in digraphs.keys():
        
        # Save the nodes for calculating in- and out-degrees later.
        degrees_helper[pathway_id].append(a)
        
        for i in list(digraphs[a].nodes):

            # Go over every descendant of every node.
            for j in nx.descendants(digraphs[a], i):

                # Find every path to this descendant.
                for path in nx.all_simple_paths(digraphs[a], i, j):
                    
                    # print(path)
                    
                    # Only use keys that we have...
                    try:
                        
                        # Get all entities for each part of the path.
                        entities = [entity_helper_copy[q] for q in path]

                        # print(entities)
                        
                        # Cartesian product of the entities.
                        # Source: https://stackoverflow.com/a/798893
                        entities = list(itertools.product(*entities))
                        
                        # print(entities)

                        # Create  a list which will hold the relations.
                        rltns = []

                        for y in range(0, len(path) - 1):

                            # Get the relations between nodes.
                            for z in list(G.edges(data=True)):

                                if z[0] == path[y] and z[1] == path[y + 1]:

                                    # Add the relation.
                                    rltns.append(z[2]['relation'])
                                    break

                        # print(rltns)
                        
                        # Make a causality chain for each possible combination
                        # of entities from above.
                        for entity_set in entities:

                            # Grow a list to make the nodes and relations.
                            # Source: https://stackoverflow.com/a/21482016
                            causal_temp = [x for x in itertools.chain.from_iterable(itertools.zip_longest(list(entity_set), rltns)) if x]

                            # Save the causality diagram.
                            causality_diagrams[pathway_id]['digraph'] = digraphs.copy()
                            causality_diagrams[pathway_id]['causality_diagrams'].append(causal_temp)
                            
                            #print(path)
                            # print(causal_temp)
                            # print('############')
                    
                    except KeyError:
                        
                        pass
    
    # print(causality_chains)
                    # print('##############################')
    
    # print(nx.is_directed_acyclic_graph(digraphs['103-31-65']))
    # print(h)
```

```python
# Go over each pathway and get the degrees.
degrees = {}

for pathway_id, info in causality_diagrams.items():
    
    # print(pathway_id)
    
    # Set the pathway ID.
    degrees[pathway_id] = {}
    
    # Now go over the causality diagrams.
    for cd in info['causality_diagrams']:
        
        # print(cd)
        
        # How long is the diagram?
        diagram_length = len(cd)
        
        # Create the integers to iterate over
        # the diagram.
        for index in range(0, diagram_length - 2, 2):
            # print(index)
            
            # Select the source, the relation,
            # and the target.
            source = cd[index]
            relation = cd[index + 1]
            target = cd[index + 2]
            
            # print(source)
            # print(relation)
            # print(target)
            
            # Save the info, assuming relations for 
            # both the source and target, even though
            # this may not be true once the entire
            # pathway has been looked at.
            if source not in degrees[pathway_id]:
                degrees[pathway_id][source] = {}
                degrees[pathway_id][source]['out'] = {}
                degrees[pathway_id][source]['out'][relation] = 0
                degrees[pathway_id][source]['in'] = {}
                degrees[pathway_id][source]['in'][relation] = 0
            
            if target not in degrees[pathway_id]:
                degrees[pathway_id][target] = {}
                degrees[pathway_id][target]['in'] = {}
                degrees[pathway_id][target]['in'][relation] = 0
                degrees[pathway_id][target]['out'] = {}
                degrees[pathway_id][target]['out'][relation] = 0
            
            # print(degrees[pathway_id])
            if relation not in degrees[pathway_id][source]['out']:
                degrees[pathway_id][source]['out'][relation] = 0
            
            if relation not in degrees[pathway_id][source]['in']:
                degrees[pathway_id][source]['in'][relation] = 0
            
            if relation not in degrees[pathway_id][target]['out']:
                degrees[pathway_id][target]['out'][relation] = 0
            
            if relation not in degrees[pathway_id][target]['in']:
                degrees[pathway_id][target]['in'][relation] = 0
            
            degrees[pathway_id][source]['out'][relation] += 1
            degrees[pathway_id][target]['in'][relation] += 1
            
            # print(degrees[pathway_id])
            
            # print('#############')
            
            
            
        # print('=============')
    
    # print(r) 
```

We can now use the degrees helper, digraphs, and mapping to entities to get the in- and out-degrees.

```python
# True degrees.
true_degrees = {}
```

```python
for pathway_id, info in causality_diagrams.items():
    
    # print(pathway_id)
    
    # Set the initial key.
    true_degrees[pathway_id] = {}
    
    if info['digraph'] != '':
        
        # print(info['digraph'])
        
        for k, v in info['digraph'].items():
            
            # print(k)
            
            for edge_tuple in v.edges:
                
                # Similar logic to above...
                
                # print(v.edges[edge_tuple])
                
                # Select the source, the relation,
                # and the target.
                source = edge_tuple[0]
                relation = v.edges[edge_tuple]['relation']
                target = edge_tuple[1]
                
                # Translate the source and the target into
                # actual gene identifiers.
                
                # Skip entities that aren't defined, such as
                # drugs and compounds.
                try:
                
                    # Get all entities for the source and target.
                    sources = entity_helper_copy[source]
                    targets = entity_helper_copy[target]

                    # Cartesian product of the entities.
                    # Source: https://stackoverflow.com/a/798893
                    entities = list(itertools.product(*[sources, targets]))

#                     print(source)
#                     print(relation)
#                     print(target)

#                     print(sources)
#                     print(targets)

#                     print('#########')
#                     print(entities)
#                     print('@@@@@@@@@@@@@')
                    
                    # Now go over the sources and targets.
                    del source
                    del target
                    
                    for tpl in entities:
                        
                        source = tpl[0]
                        target = tpl[1]

                        # Save the info, assuming relations for 
                        # both the source and target, even though
                        # this may not be true once the entire
                        # pathway has been looked at.
                        if source not in true_degrees[pathway_id]:
                            true_degrees[pathway_id][source] = {}
                            true_degrees[pathway_id][source]['out'] = {}
                            true_degrees[pathway_id][source]['out'][relation] = 0
                            true_degrees[pathway_id][source]['in'] = {}
                            true_degrees[pathway_id][source]['in'][relation] = 0

                        if target not in true_degrees[pathway_id]:
                            true_degrees[pathway_id][target] = {}
                            true_degrees[pathway_id][target]['in'] = {}
                            true_degrees[pathway_id][target]['in'][relation] = 0
                            true_degrees[pathway_id][target]['out'] = {}
                            true_degrees[pathway_id][target]['out'][relation] = 0

                        if relation not in true_degrees[pathway_id][source]['out']:
                            true_degrees[pathway_id][source]['out'][relation] = 0

                        if relation not in true_degrees[pathway_id][source]['in']:
                            true_degrees[pathway_id][source]['in'][relation] = 0

                        if relation not in true_degrees[pathway_id][target]['out']:
                            true_degrees[pathway_id][target]['out'][relation] = 0

                        if relation not in true_degrees[pathway_id][target]['in']:
                            true_degrees[pathway_id][target]['in'][relation] = 0

                        true_degrees[pathway_id][source]['out'][relation] += 1
                        true_degrees[pathway_id][target]['in'][relation] += 1

                except KeyError:

                    pass
            
# print(yyyyyy)
```
